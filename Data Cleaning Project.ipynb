```python
# =========================================
# 📌 Data Cleaning Pipeline
# =========================================

pipeline_description = """
## 📌 Data Cleaning Pipeline

### 🎯 Objective
Prepare and standardize the dataset to ensure data quality, consistency,
and readiness for exploratory analysis and modeling.

### 🔧 Steps Performed

1. Deleting redundant columns
2. Renaming the columns
3. Dropping duplicates
4. Removing NaN values from the dataset
5. Cleaning and formatting individual columns
6. Checking for additional transformations
"""

print(pipeline_description)
```

```python
# ==============================
# STEP 0.1: Load dataset
# ==============================

print("STEP 0.1: Loading Airbnb dataset")

import pandas as pd

data = pd.read_csv("Airbnb_Open_Data.csv")

print("Dataset loaded successfully.")

# ==============================
# STEP 0.2: Preview data
# ==============================

print("\nSTEP 0.2: Displaying first 10 rows")
print(data.head(10))
```

```python
# ==============================
# STEP 0.3: Inspect dataset structure
# ==============================

print("\nSTEP 0.3.1: Reviewing dataset columns")
print(data.columns)

print("\nSTEP 0.3.2: Dataset info")
data.info()
```

```python
# ==============================
# STEP 1.0: Decide which columns to analyze
# ==============================

print("STEP 1: Selecting relevant columns for analysis")

columns_to_keep = [
    'NAME', 'host id', 'host_identity_verified', 'host name',
    'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
    'country code', 'instant_bookable', 'cancellation_policy', 'room type',
    'Construction year', 'price', 'service fee', 'minimum nights',
    'number of reviews', 'last review'
]

columns_to_drop = [
    'id', 'reviews per month', 'review rate number',
    'calculated host listings count', 'availability 365',
    'house_rules', 'license'
]

print("Number of columns to KEEP:", len(columns_to_keep))
print("Number of columns to DROP:", len(columns_to_drop))


# ==============================
# STEP 1.1: Apply column filtering
# ==============================

# Keep only selected columns
data = data[columns_to_keep].copy()

# Or alternatively drop unwanted columns:
# data_filtered = data.drop(columns=columns_to_drop)


# ==============================
# STEP 1.2: Verify result
# ==============================

print("\nFinal number of columns in dataset:", data.shape[1])
print("Columns in the filtered dataset:")
print(data.columns.tolist())
```

```python
# ==============================
# STEP 2.0: Standardize column names
# ==============================

print("STEP 2.0: Converting column names to uppercase")

new_columns_names = []
for i in data.columns:
    new_columns_names.append(i.upper())

data.columns = new_columns_names

print("Updated column names:")
print(data.columns.tolist())
```

```python
# ==============================
# STEP 3: Drop duplicate rows
# ==============================

print("STEP 3: Removing duplicate rows")

initial_rows = data.shape[0]

data = data.drop_duplicates().copy()

final_rows = data.shape[0]
duplicates_removed = initial_rows - final_rows

print("Initial number of rows:", initial_rows)
print("Final number of rows:", final_rows)
print("Number of duplicates removed:", duplicates_removed)# ==============================
# STEP 4: Clean LAST REVIEW and handle NaN values carefully
# ==============================

print("STEP 4: Removing duplicates based on LAST REVIEW and handling NaN values")

# --- Part A: Drop duplicates ONLY using LAST REVIEW ---
initial_rows = data.shape[0]

data = data.drop_duplicates(subset=['LAST REVIEW']).copy()

remaining_rows = data.shape[0]
dropped_rows = initial_rows - remaining_rows

print("Initial number of rows:", initial_rows)
print("Number of rows removed (duplicates on LAST REVIEW):", dropped_rows)
print("Number of rows remaining:", remaining_rows)
```

```python
# ==============================
# STEP 4: Remove NaN values (targeted)
# ==============================

print("STEP 4: Removing rows where LAST REVIEW or NAME if is NaN")

initial_rows = data.shape[0]

# Remove rows with NaN only in the specified columns
data = data.dropna(subset=['LAST REVIEW', 'NAME']).copy()

remaining_rows = data.shape[0]
dropped_rows = initial_rows - remaining_rows

print("Initial number of rows:", initial_rows)
print("Number of rows removed due to NaN:", dropped_rows)
print("Number of rows remaining:", remaining_rows)

# Note for documentation
print("\nNOTE: If we were to remove ALL NaN values from the entire dataset,")
print("the dataframe would be reduced to only one row.")
```

```python
# ==============================
# STEP 5: Cleaning and formatting individual columns
# ==============================

print("STEP 5: Cleaning and formatting individual columns")

# =====================================================
# 5.1 Encode HOST_IDENTITY_VERIFIED → binary
# =====================================================
print("\n[5.1] Encoding HOST_IDENTITY_VERIFIED (1 = verified, 0 = unconfirmed)")

print("Unique values BEFORE encoding:")
print(data['HOST_IDENTITY_VERIFIED'].unique())

verification_map = {
    'verified': 1,
    'unconfirmed': 0,
    't': 1,
    'f': 0,
    'yes': 1,
    'no': 0,
    True: 1,
    False: 0
}

data['HOST_IDENTITY_VERIFIED'] = (
    data['HOST_IDENTITY_VERIFIED']
    .astype(str)
    .str.strip()
    .str.lower()
    .map(verification_map)
)

print("Unique values AFTER encoding:")
print(data['HOST_IDENTITY_VERIFIED'].unique())


# =====================================================
# 5.2 Encode CANCELLATION_POLICY → ordinal
# =====================================================
print("\n[5.2] Encoding CANCELLATION_POLICY (flexible=1, moderate=2, strict=3)")

print("Unique values BEFORE encoding:")
print(data['CANCELLATION_POLICY'].unique())

cancellation_map = {
    'flexible': 1,
    'moderate': 2,
    'strict': 3
}

data['CANCELLATION_POLICY'] = (
    data['CANCELLATION_POLICY']
    .astype(str)
    .str.strip()
    .str.lower()
    .map(cancellation_map)
)

print("Unique values AFTER encoding:")
print(data['CANCELLATION_POLICY'].unique())


# =====================================================
# 5.3 Encode INSTANT_BOOKABLE → binary
# =====================================================
print("\n[5.3] Encoding INSTANT_BOOKABLE (1 = True, 0 = False)")

print("Unique values BEFORE encoding:")
print(data['INSTANT_BOOKABLE'].unique())

instant_map = {
    'true': 1,
    'false': 0,
    't': 1,
    'f': 0,
    True: 1,
    False: 0
}

data['INSTANT_BOOKABLE'] = (
    data['INSTANT_BOOKABLE']
    .astype(str)
    .str.strip()
    .str.lower()
    .map(instant_map)
)

print("Unique values AFTER encoding:")
print(data['INSTANT_BOOKABLE'].unique())

# =====================================================
# 5.4 Remove dollar sign and convert monetary columns
# =====================================================
print("\n[5.4] Cleaning PRICE and SERVICE FEE (remove $, commas, and convert to float)")

money_columns = ['PRICE', 'SERVICE FEE']

for col in money_columns:
    print(f"\nProcessing column: {col}")
    print("Sample BEFORE cleaning:")
    print(data[col].head(3))

    data[col] = (
        data[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)  # ⭐ FIX: remove thousands separator
        .str.strip()
    )

    # Safe numeric conversion
    data[col] = pd.to_numeric(data[col], errors='coerce')

    print("Sample AFTER cleaning:")
    print(data[col].head(3))

print("\n✅ STEP 5 completed: Monetary columns cleaned successfully.")
```

```python
# ==============================
# STEP 6: Checking for additional transformations
# ==============================

print("STEP 6: Checking for additional transformations")

# ---------------------------------
# 6.1 Check data types
# ---------------------------------
print("\n[6.1] Reviewing data types")
print(data.dtypes)

# ---------------------------------
# 6.2 Check for remaining missing values
# ---------------------------------
print("\n[6.2] Checking remaining missing values")
missing_values = data.isna().sum()
print(missing_values[missing_values > 0])

# ---------------------------------
# 6.3 Basic statistical summary
# ---------------------------------
print("\n[6.3] Generating statistical summary")
print(data.describe())

# ---------------------------------
# 6.4 Check for potential outliers in numeric columns
# ---------------------------------
print("\n[6.4] Checking numeric column ranges")

numeric_cols = data.select_dtypes(include='number').columns

for col in numeric_cols:
    print(f"{col} → min: {data[col].min()}, max: {data[col].max()}")

print("\n✅ STEP 6 completed: Additional transformation review finished.")
```
